1 What is Simple Linear Regression
It is a statistical method used to model the relationship between two variables: one independent variable (X) and one dependent variable (Y), by fitting a linear equation to observed data.
2  What are the key assumptions of Simple Linear Regression?
Linearity: The relationship between X and Y is linear. Independence: Observations are independent of each other. Homoscedasticity: The variance of residual errors is constant across all levels of X. Normality: The residuals (errors) are normally distributed.
3 What does the coefficient m represent in the equation Y=mX+c?
The slope. It represents the change in y for every one-unit change in x. It indicates the direction and strength of the relationship.
4  What does the intercept c represent in the equation Y=mX+c
The y-intercept. It is the predicted value of Y when X = 0.
5 How do we calculate the slope m in Simple Linear Regression
m is calculated using y predicted and y actual and x actual and x predicted, by using the formula y actual - y predicted / x actual - x predicted
6  What is the purpose of the least squares method in Simple Linear Regression
To find the line that minimizes the sum of the squares of the vertical deviations (residuals) between each data point and the fitted line. This ensures the best fit.
7 How is the coefficient of determination (R²) interpreted in Simple Linear Regression
It represents the proportion of the variance in the dependent variable that is predictable from the independent variable. An $R^2$ of 0.80 means 80% of the variation in y is explained by x
8 What is Multiple Linear Regression
An extension of SLR used to predict the value of a dependent variable based on two or more independent variables.
9 What is the main difference between Simple and Multiple Linear Regression
 The number of independent variables: SLR has one x, while MLR has multiple (x1 , x2 , till xn).
10  What are the key assumptions of Multiple Linear Regression
Same as SLR, plus No Multicollinearity: Independent variables should not be highly correlated with each other.
11  What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model
It occurs when the variance of the residuals is not constant. It makes standard errors unreliable, which can lead to incorrect conclusions about the significance of predictors.
12 How can you improve a Multiple Linear Regression model with high multicollinearity
Remove highly correlated independent variables. Use Principal Component Analysis (PCA) to combine variables. Use Ridge or Lasso regression.
13  What are some common techniques for transforming categorical variables for use in regression models
One-Hot Encoding: Creating dummy binary variables (0 or 1).As there are 3 variables in target variable then 3 models are created like the variable in this model are 0,1,2. 0 will be given as 1 in one model and rest 2 are treated as 0 and in next 1 will be treated as 1 and rest 2 will be treated as 0 and so on.
14 What is the role of interaction terms in Multiple Linear Regression
 They account for situations where the effect of one independent variable on y changes depending on the level of another independent variable eg x1 and x2
15 How can the interpretation of intercept differ between Simple and Multiple Linear Regression
In SLR, it's Y when X=0. In MLR, it is the value of Y when all independent variables are zero.
16 What is the significance of the slope in regression analysis, and how does it affect predictions
 It provides a "baseline" value. However, if X=0 is outside the range of your data (e.g., a person with 0 height), the intercept may only serve as a mathematical anchor rather than a physical reality.
17 How does the intercept in a regression model provide context for the relationship between variables
 It provides a "baseline" value. However, if X=0 is outside the range of your data (e.g., a person with 0 height), the intercept may only serve as a mathematical anchor rather than a physical reality.
18 What are the limitations of using R² as a sole measure of model performance
It always increases as you add more variables, even if they are irrelevant. It doesn't indicate if the model is biased. It doesn't tell you if you have chosen the right regression type (linear vs. non-linear).
19  How would you interpret a large standard error for a regression coefficient
This is a red flag for overfitting. It means you have added variables that do not contribute significantly to the model's predictive power.
20 How can heteroscedasticity be identified in residual plots, and why is it important to address it
Heteroscedasticity is identified by a "fan" or "cone" pattern in a residual plot, where the vertical spread of the dots grows wider or narrower as you move across the x-axis. A healthy model should show a random, uniform cloud of points.It must be addressed because it makes your p-values and confidence intervals unreliable. While the model may still predict the average correctly, it loses its ability to accurately measure the significance of its predictors, often leading to "false positives" in your results.
21  What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?
It means you have added too many irrelevant independent variables to your model. While r2 will always increase (or stay the same) as you add more predictors—even if they are just random noise—the Adjusted r2 penalizes you for adding variables that don't actually improve the model's predictive power.A large gap between the two suggests your model is "memorizing" the noise in your specific dataset rather than finding a true relationship that will work on new, unseen data.
22  Why is it important to scale variables in Multiple Linear Regression
A high r2 with a low adjusted r2 is a classic sign of overfitting.It means your model includes too many irrelevant variables that don't actually help predict the outcome. While r2 automatically increases every time you add a new variable (even if it's just "noise"), Adjusted r2 only increases if that variable significantly improves the model.A large gap between the two suggests your model is "memorizing" your specific data instead of finding a real, generalizable pattern.
23 What is polynomial regression
Polynomial regression is a form of regression analysis used when the relationship between the independent variable x and the dependent variable y is non-linear.Instead of fitting a straight line, it fits a curve to the data points. It does this by taking the original features and raising them to a power (e.g., x^2, x^3), effectively turning a non-linear problem into a linear one that the model can solve.
24  How does polynomial regression differ from linear regression
he primary difference lies in the shape of the relationship they model between variables:Linear Regression fits a straight line (y = mx + c). It assumes that as $X$ increases, Y changes at a constant rate.Polynomial Regression fits a curve. It allows the model to capture more complex relationships—like data that goes up and then down—by adding exponents to the independent variables (e.g., x^2, x^3).
25 When is polynomial regression used
Polynomial regression is used when the relationship between variables is non-linear, meaning a straight line fails to capture the true trend of the data. It is ideal for modeling "curved" patterns, such as the trajectory of a projectile, population growth that eventually levels off, or economic scenarios with diminishing returns. By adding exponents to the independent variables, the model creates a flexible curve that can follow the data's natural shape more accurately than a linear model.
26  What is the general equation for polynomial regression
y = b0 + b1x + b2x**2 + bnx**n + e
y : Dependent variable (target).
x :  Independent variable (feature)
b0 : Intercept
b1.....bn: Coefficients for linear, quadratic, cubic, up to
-th degree terms.
e : residual error
27 Can polynomial regression be applied to multiple variables
Yes, it can. When you apply polynomial regression to multiple independent variables, it is often called Multiple Polynomial Regression.In this case, the model doesn't just include the powers of each variable , but also interaction terms. This allows the model to capture how the variables affect each other.
28  What are the limitations of polynomial regression
Polynomial regression has several key limitations, primarily its high risk of overfitting. When the degree of the polynomial is too high, the curve wiggles to capture random noise in the training data rather than the true underlying trend, leading to poor performance on new data. Additionally, polynomials are notoriously unreliable for extrapolation; the curve can behave erratically and shoot toward infinity just outside the range of the original data points. Finally, higher-order terms make the model extremely sensitive to outliers and much harder to interpret than a simple linear relationship.
29  What methods can be used to evaluate model fit when selecting the degree of a polynomial
To evaluate the best degree for a polynomial, the most effective method is Cross-Validation, where you compare the error rates of different degrees on a separate validation dataset. You look for the "sweet spot" where the Validation Error is at its lowest; if the error on the training data continues to drop while the validation error begins to rise, the model is overfitting.Other key metrics include Adjusted $R^2$, which penalizes adding unnecessary polynomial terms that don't truly improve the model, and Information Criteria (AIC/BIC), which score the model based on a balance of accuracy and simplicity. Finally, performing a Residual Analysis can reveal if a pattern still exists in the errors, suggesting a higher degree is needed, or if the errors are a random cloud, indicating a good fit.
30 Why is visualization important in polynomial regression
Visualization is essential in polynomial regression because it is the most intuitive way to detect the Bias-Variance Tradeoff. Since polynomial models can become incredibly complex, simply looking at numerical scores like r2 can be misleading.
